```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des données
data = pd.read_csv("Camp_Market.csv", sep=';')

# Encodage One-Hot pour les colonnes 'Education' et 'Marital_Status'
data_encoded = pd.get_dummies(data, columns=['Education', 'Marital_Status'])

# Suppression de la colonne 'Dt_Customer'
data_encoded = data_encoded.drop(columns=['Dt_Customer'])

# Calcul de la matrice de corrélation pour le dataframe encodé
corr_matrix_encoded = data_encoded.corr()

# Visualisation de la matrice de corrélation
plt.figure(figsize=(15, 13))
sns.heatmap(corr_matrix_encoded, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Matrice de Corrélation")
plt.show()
```
